## Research

### IBCS

IBCS (International Business Communication Standards) is a set of guidelines for designing reports, presentations, and dashboards in business contexts. The primary goal of IBCS is to improve the comprehensibility and effectiveness of business communications by promoting a uniform structure and design. ​(Trusteddecisions, 2023)​
The core of IBCS is the SUCCESS formula, which includes principles like Simplify, Unify, Condense, Check, Enable, Structure, and Show ​(McCoy, 2018)​.  These principles guide the creation of standardized and impactful dashboards, focusing on aspects such as proper visualization, logical structure, and effective content organization ​(Damen, n.d.)​.

IBCS standards are widely adopted in various areas of corporate management, including financial reporting, sales analysis, and strategic management. Many large corporations, such as SAP and Philips, have successfully implemented IBCS in their reporting practices ​(Damen, n.d.)​.


### IBCS Rules

In the context of developing an AI-driven solution for assessing IBCS compliance in dashboards, it is crucial to understand the specific rules and guidelines that form the foundation of IBCS standards. This research focuses on four key rules: UN 3.3, UN 3.4, SI 2.2, and UN 4.1.

Rules UN 3.3 and UN 3.4 fall under the "Unify" principle of IBCS and address the orientation of charts. These guidelines stipulate that time periods should always be displayed horizontally, typically on the x-axis, aligning with the natural perception of time progressing from left to right. Conversely, structure dimensions (such as product categories or regions) should be displayed vertically, usually on the y-axis. This standardization enhances readability and improves comparability across various reports and dashboards ​(IBCS, n.d.)​.

Rules SI 2.2 and UN 4.1 focus on color usage in business communications. SI 2.2, part of the "Simplify" principle, advises against the use of decorative or unnecessary colors, emphasizing that colors should only be employed when they convey specific meaning. UN 4.1, under the "Unify" principle, stresses the importance of using consistent colors for scenario analyses. For instance, actual data (AC) is typically represented in solid black, plan data (PL) with outlined elements, and forecast data (FC) with hatched elements. Furthermore, IBCS recommends specific colors for variances: red for negative and green for positive.

These color conventions facilitate quick identification of different scenarios and variances without the need for legends or additional explanations. By adhering to these guidelines, chart creators can minimize bias and enhance clarity in business communications, allowing decision-makers to focus on the data rather than deciphering inconsistent visual representations ​(Hichter, 2015)​.

## 2. Load and Preprocess the Data (Traditional ML)

We load images from the `IBCS_COLOR` folder in Google Drive:
- Subfolders (`Compliant`, `NonCompliant`) provide labels.
- Images are resized to 256x256 (RGB)
- Each image is flattened into a 1D array for input into traditional ML models

*Why this method?*  
Flattened pixel values let us quickly test models like SVM without needing complex feature extraction.

```python
def load_image_from_drive(file_id, size, flatten=True):
    """Download and preprocess an image from Google Drive."""
    request = drive_service.files().get_media(fileId=file_id)
    fh = io.BytesIO()
    downloader = MediaIoBaseDownload(fh, request)
    done = False
    while done is False:
        status, done = downloader.next_chunk()
    fh.seek(0)
    img = Image.open(fh).convert('RGB')
    img = img.resize((size, size))
    img_array = np.array(img)
    if flatten:
        return img_array.flatten()
    return img_array

def load_labelled_images_from_drive(folder_id, size, flatten=True):
    """Load images from Google Drive folder and assign labels."""
    labels, images = [], []
    
    query = f"'{folder_id}' in parents and mimeType = 'application/vnd.google-apps.folder'"
    page_token = None
    subfolders = []
    while True:
        response = drive_service.files().list(
            q=query,
            fields='nextPageToken, files(id, name)',
            pageToken=page_token
        ).execute()
        subfolders.extend(response.get('files', []))
        page_token = response.get('nextPageToken')
        if not page_token:
            break
    
    print(f"Found {len(subfolders)} subfolders: {[f['name'] for f in subfolders]}")
    
    for subfolder in subfolders:
        label = subfolder['name']
        subfolder_id = subfolder['id']
        
        query = f"'{subfolder_id}' in parents and (mimeType = 'image/jpeg' or mimeType = 'image/png')"
        page_token = None
        files = []
        while True:
            response = drive_service.files().list(
                q=query,
                fields='nextPageToken, files(id, name, mimeType)',
                pageToken=page_token
            ).execute()
            files.extend(response.get('files', []))
            page_token = response.get('nextPageToken')
            if not page_token:
                break
        
        print(f"Found {len(files)} files in subfolder '{label}'")
        
        for file in files:
            try:
                img = load_image_from_drive(file['id'], size, flatten)
                images.append(img)
                labels.append(label)
            except Exception as e:
                print(f"Error processing {file['name']} in {label}: {e}")
    
    if not images:
        print("⚠️ No images found.")
        return np.array([]), np.array([])
    
    print(f"Total images loaded: {len(images)}")
    return np.array(images), np.array(labels)

# Load images for traditional ML
images_ml, labels_ml = load_labelled_images_from_drive(FOLDER_ID, SIZE, flatten=True)
```

## 5. Train and Compare SVM Models

We trained three SVM models using different kernels to compare performance.

### 🔹 Default SVM (RBF kernel): Non-linear kernel for complex boundaries.

```python
# Default SVM (RBF)
model_default = SVC()
model_default.fit(X_train_ml, y_train_ml)
y_pred = model_default.predict(X_test_ml)
print("Accuracy SVM (RBF):", model_default.score(X_test_ml, y_test_ml))
print(classification_report(y_test_ml, y_pred, target_names=label_encoder.classes_))
```

### 🔹 Linear SVM: Assumes color distributions are linearly separable.

```python
# Linear SVM
model_linear = SVC(kernel='linear')
model_linear.fit(X_train_ml, y_train_ml)
y_pred = model_linear.predict(X_test_ml)
print("Accuracy SVM (Linear):", model_linear.score(X_test_ml, y_test_ml))
print(classification_report(y_test_ml, y_pred, target_names=label_encoder.classes_))
```

### 🔹 Polynomial SVM: Tests curved decision boundaries.

```python
# Polynomial SVM
model_poly = SVC(kernel='poly')
model_poly.fit(X_train_ml, y_train_ml)
y_pred = model_poly.predict(X_test_ml)
print("Accuracy SVM (Polynomial):", model_poly.score(X_test_ml, y_test_ml))
print(classification_report(y_test_ml, y_pred, target_names=label_encoder.classes_))
```

**Observation:**

## 7. Tune the Polynomial SVM (GridSearchCV)
Even though the linear kernel performed best at first, we want to test if a more flexible polynomial model could improve with the **right configuration**. To improve performance, we tuned the polynomial SVM using **GridSearchCV** over these hyperparameters:

- `C`: [0.1, 1, 10]  
- `degree`: [2, 3, 4]  
- `gamma`: ['scale', 'auto']  
- `coef0`: [0.0, 1.0]

## 10. Deep Learning with ResNet50
To overcome the limitations of traditional models, we implement a ResNet50 CNN, which automatically extracts color and pattern features for IBCS Rule B compliance. ResNet50 is chosen for its efficiency and proven performance in image classification (Krizhevsky et al., 2017).

### 10.1 Load and Preprocess Data for CNN
We reload images in 256x256x3 format (not flattened), apply ResNet50 preprocessing (ImageNet mean subtraction), and normalize pixel values to [0, 1].

```python
from tensorflow.keras.applications.resnet50 import preprocess_input

# Load images for CNN (non-flattened)
images_cnn, labels_cnn = load_labelled_images_from_drive(FOLDER_ID, SIZE, flatten=False)
labels_encoded_cnn = label_encoder.fit_transform(labels_cnn)

# Normalize images and apply ResNet50 preprocessing
images_cnn = images_cnn / 255.0
images_cnn = preprocess_input(images_cnn * 255.0)  # ResNet50 preprocessing

# Split dataset (80% train, 20% validation)
X_train_cnn, X_val_cnn, y_train_cnn, y_val_cnn = train_test_split(
    images_cnn, labels_encoded_cnn, test_size=0.2, random_state=42
)
```

### 10.2 Precompute Data Augmentation
To speed up training and improve generalization, we precompute augmented images:
- Generate 3 augmented versions per image.
- Apply rotation (±20°), flips, zoom (±10%), brightness adjustment (±20%), color jittering, and shear (±10%).
- This expands the training set from ~361 to ~1444 images (361 × 4).

```python
datagen = ImageDataGenerator(
    rotation_range=20,
    horizontal_flip=True,
    vertical_flip=True,
    zoom_range=0.1,
    brightness_range=[0.8, 1.2],
    channel_shift_range=20,  # Color jittering
    shear_range=0.1  # Shear ±10%
)

# Precompute augmented images
X_train_aug = []
y_train_aug = []
augmentations_per_image = 3

for i in range(len(X_train_cnn)):
    img = X_train_cnn[i]
    label = y_train_cnn[i]
    # Add original image
    X_train_aug.append(img)
    y_train_aug.append(label)
    # Generate augmented images
    img = img.reshape((1,) + img.shape)  # Reshape for generator
    for _ in range(augmentations_per_image):
        aug_img = next(datagen.flow(img, batch_size=1))[0]
        X_train_aug.append(aug_img)
        y_train_aug.append(label)

X_train_aug = np.array(X_train_aug)
y_train_aug = np.array(y_train_aug)

print(f"Augmented training set size: {len(X_train_aug)} images")
```

### 10.3 Build and Fine-Tune ResNet50
We use a pre-trained ResNet50 (ImageNet weights), add a custom classification head, and fine-tune the model with the following parameters:

- Learning Rate: 0.0001 (for stable convergence).
- Batch Size: 32 (suitable for CPU training).
- Epochs: 50 (with early stopping, patience=10).
- Optimizer: Adam.
- Loss Function: Binary cross-entropy.
- Class Weights: To handle minor class imbalance.
- We also unfreeze the last 10 layers of ResNet50 to allow fine-tuning.

**How ResNet50 Works in This Project**

ResNet50 is a deep convolutional neural network (CNN) with 50 layers, pre-trained on ImageNet (1.2M images, 1000 classes). It uses residual connections (skip connections) to address the vanishing gradient problem, allowing deeper networks to train effectively (He et al., 2016). Here’s how it works for IBCS Rule B compliance:
- **Feature Extraction**: ResNet50 processes input images (256x256x3) through convolutional layers, extracting hierarchical features like edges, textures, and color patterns. For IBCS Rule B, it learns to detect specific color usage (e.g., green for positive, red for negative) and identify decorative or inconsistent colors.
- **Transfer Learning**: We use pre-trained ImageNet weights to leverage general image features (e.g., color and texture detection), then fine-tune the last 10 layers to adapt to our task. This balances learning task-specific features (IBCS color rules) with retaining general image understanding.
- **Custom Head**: After the base ResNet50, we add:
    - **GlobalAveragePooling2D**: Reduces spatial dimensions (e.g., 8x8x2048 to 2048) by averaging, summarizing features.
    - **Dense(128, relu)**: A fully connected layer with 128 neurons to learn task-specific patterns.
    - **Dropout(0.5)**: Randomly drops 50% of neurons during training to prevent overfitting.
    - **Dense(1, sigmoid)**: Outputs a probability (0 to 1) for binary classification (0 = Compliant, 1 = NonCompliant).
- **Training**: We use:
    - **Adam Optimizer**: Adapts the learning rate (0.0001) for faster convergence.
    - **Binary Cross-Entropy Loss**: Measures the difference between predicted probabilities and true labels.
    - **Class Weights**: Balance the model’s predictions for Compliant and NonCompliant classes.
    - **Early Stopping**: Stops training if validation loss doesn’t improve for 10 epochs, preventing overfitting.
- **Fine-Tuning**: Unfreezing the last 10 layers allows ResNet50 to adjust its filters to better detect IBCS color patterns (e.g., green/red usage).

```python
from tensorflow.keras.layers import Dropout
from sklearn.utils.class_weight import compute_class_weight

# Load ResNet50 with ImageNet weights
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(SIZE, SIZE, 3))

# Unfreeze the last 10 layers for fine-tuning
base_model.trainable = True
for layer in base_model.layers[:-10]:
    layer.trainable = False

# Add custom head with dropout
inputs = Input(shape=(SIZE, SIZE, 3))
x = base_model(inputs, training=False)
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)  # Add dropout to reduce overfitting
outputs = Dense(1, activation='sigmoid')(x)
model = Model(inputs, outputs)

# Compute class weights to handle bias
class_weights = compute_class_weight('balanced', classes=np.unique(y_train_aug), y=y_train_aug)
class_weight_dict = {0: class_weights[0], 1: class_weights[1]}

# Compile model
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Define early stopping
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

# Train model with batch size of 32
history = model.fit(
    X_train_aug, y_train_aug,  # Use precomputed augmented data
    batch_size=32,  # Batch size: 32
    epochs=50,
    validation_data=(X_val_cnn, y_val_cnn),
    callbacks=[early_stopping],
    class_weight=class_weight_dict
)
```

### 10.4 Evaluate ResNet50
We evaluate the model on the validation set and visualize the confusion matrix.

```python
# Predict on validation set
y_pred_cnn = (model.predict(X_val_cnn) > 0.5).astype(int).flatten()

# Confusion Matrix
cm = confusion_matrix(y_val_cnn, y_pred_cnn)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=label_encoder.classes_,
            yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix - ResNet50")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Classification Report
print("Accuracy ResNet50:", accuracy_score(y_val_cnn, y_pred_cnn))
print(classification_report(y_val_cnn, y_pred_cnn, target_names=label_encoder.classes_))

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()
```

- **Error Analysis**: The model’s errors (2 False Negatives, 1 False Positive) suggest it still struggles in some cases.
- **Training Stability**: The model converges quickly (train accuracy reaches 1.0 by Epoch 4), but validation metrics fluctuate due to the small validation set (91 images). Early stopping effectively prevents overfitting by selecting weights from Epoch 2.
- **Generalization**: The small train-val accuracy gap (1.0 vs. 0.967) indicates good generalization, though the validation loss fluctuations suggest room for improvement, possibly by increasing the validation set size or adding more diverse training data.

These observations confirm that the ResNet50 model is highly effective for IBCS Rule B compliance detection, but minor adjustments to the training data could further improve its robustness for edge cases.
